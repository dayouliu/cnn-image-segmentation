#### Example 1

This class takes a img, target pair, and then transform the pair such that they are in `Torch.Tensor()` format.

#### Solution:

```python
class JointToTensor(object):
    def __call__(self, img, target):
        return tF.to_tensor(img), torch.from_numpy(np.array(target.convert('P'), dtype=np.int32)).long()
```

#### Example 2: 

This class implements CenterCrop that takes an img, target pair, and then apply a crop about the center of the image such that the output resolution is $\mbox{size} \times \mbox{size}$.

#### Solution:

```python
class JointCenterCrop(object):
    def __init__(self, size):
        """
        params:
            size (int) : size of the center crop
        """
        self.size = size
        
    def __call__(self, img, target):
        return (tF.five_crop(img, self.size)[4], 
                tF.five_crop(target, self.size)[4])
    
img, target = JointCenterCrop(100)(*sample1)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

#### (a) Implement RandomFlip

This class should take a img, target pair and then apply a horizontal flip across the vertical axis at random.

#### Solution:

```python
class JointRandomFlip(object):
    def __call__(self, img, target):
        if random.random() < 0.5:
            return tF.hflip(img), tF.hflip(target)
        else:
            return img, target

img, target = JointRandomFlip()(*sample1)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

#### (b) Implement RandomResizeCrop

This class should take a img, target pair and then resize the images by a random scale between $[\mbox{minimum_scale}, \mbox{maximum_scale}]$, crop a random location of the image by $\min(\mbox{size}, \mbox{image_height}, \mbox{image_width})$ (where the size is passed in as an integer in the constructor), and then resize to $\mbox{size} \times \mbox{size}$ (again, the size passed in). The crop box should fit within the image.

#### Solution:

```python
class JointRandomResizeCrop(object):
    def __init__(self, size = 200, min_scale = 0.75, max_scale = 1.25):
        self.size = size
        self.min_scale = min_scale
        self.max_scale = max_scale
        
    def __call__(self, img, target):
        size = self.size
        min_scale = self.min_scale
        max_scale = self.max_scale
        
        scale = random.uniform(min_scale, max_scale)
        img = tF.resize(img, (int(img.size[0] * scale), int(img.size[1] * scale)))
        target = tF.resize(target, (int(target.size[0] * scale), int(target.size[1] * scale)))

        crop_size = int(min(size, img.size[0], img.size[1]))
        random_crop = transforms.RandomCrop.get_params(img, output_size=(crop_size, crop_size))

        img = tF.crop(img, *random_crop)
        target = tF.crop(target, *random_crop)

        img = tF.resize(img, (size, size))
        target = tF.resize(target, (size, size))
        return img, target

img, target = JointRandomResizeCrop()(*sample1)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

This class should take a img, target pair and then normalize the images by subtracting the mean and dividing variance. 

#### Solution:

```python
norm = ([0.485, 0.456, 0.406], 
        [0.229, 0.224, 0.225]) 

class JointNormalize(object):
    def __init__(self, mean, std):
        self.mean = mean
        self.std = std
        
    def __call__(self, img, target):
        img = tF.normalize(img, self.mean, self.std)
        return img, target

img, target = JointToTensor()(*sample1)
img, target = JointNormalize(norm[0], norm[1])(img, target)
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img.permute(1,2,0))    # need to permute!
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(target)
```

#### (d) Compose the transforms together: 
##### Use `JointCompose` (fully implemeted below) to compose the implemented transforms together in some random order. Verify the output makes sense and visualize it.

```python
# This class composes transofrmations from a given list of image transforms (expected in the argument). Such compositions 
# will be applied to the dataset during training. This cell is fully implemented.

class JointCompose(object):
    def __init__(self, transforms):
        """
        params: 
           transforms (list) : list of transforms
        """
        self.transforms = transforms

    # We override the __call__ function such that this class can be
    # called as a function i.e. JointCompose(transforms)(img, target)
    # Such classes are known as "functors"
    def __call__(self, img, target):
        """
        params:
            img (PIL.Image)    : input image
            target (PIL.Image) : ground truth label 
        """
        assert img.size == target.size
        for t in self.transforms:
            img, target = t(img, target)
        return img, target
```

```python
import torchvision.models as models
from torchsummary import summary

resnet34 = models.resnet34(pretrained=True)
print(resnet34)
```

```python
# change to resnet50 LATER!

class MyNet(nn.Module):
    def __init__(self, num_classes, criterion=None):
        super(MyNet, self).__init__()

        # Implement me
        self.encode1 = resnet34.conv1
        self.bn1 = resnet34.bn1
        self.relu1 = resnet34.relu
        self.maxpool1 = resnet34.maxpool
        self.encode2 = resnet34.layer1
        self.encode3 = resnet34.layer2
        self.encode4 = resnet34.layer3
        self.encode5 = resnet34.layer4

        self.decoder = nn.Sequential(
            nn.Conv2d(512+64, 256, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.Upsample(scale_factor=2),
            nn.Conv2d(256, 128, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, num_classes, kernel_size=1, stride=1, padding=0)
        )

        self.criterion = criterion
    
    def forward(self, inp, gts=None):
        # encode 
        size = inp.size()
        height = size[2]
        width = size[3]

        e1 = self.encode1(inp)
        e1 = self.bn1(e1)
        e1 = self.relu1(e1)
        e1 = self.maxpool1(e1)

        e2 = self.encode2(e1)
        size_e2 = e2.size()
        height_e2 = size_e2[2]
        width_e2 = size_e2[3]

        e3 = self.encode3(e2)
        e4 = self.encode4(e3)
        e5 = self.encode5(e4)

        u1 = F.upsample(e5, size=(height_e2, width_e2))

        c1 = torch.cat((e2, u1), 1)

        # decode
        d1 = self.decoder(c1)

        lfinal = F.upsample(d1, size=(height, width))

        # Implement me
        
        if self.training:
            # Return the loss if in training mode
            return self.criterion(lfinal, gts)              
        else:
            # Return the actual prediction otherwise
            return lfinal
```

```python
import torch.optim as optim

def get_optimizer(net):
    # Write me
    return optim.Adam(net.parameters(), lr=0.0001)
```




#### (b) Write the training loop to train the network. 


```python
def train(train_loader, net, optimizer, loss_graph):
    
    if USE_GPU:
        net = net.cuda()

    for i, data in enumerate(train_loader):
        
        inputs, masks = data

        if USE_GPU:
            inputs = inputs.cuda()
            masks = masks.cuda()
        
        # Write me
        inputs = inputs.float()
        optimizer.zero_grad()
        loss = net(inputs, gts=masks)
        loss_graph.append(loss.item()) # Populate this list to graph the loss
        loss.backward()
        optimizer.step()
    

    return loss
```

#### For the network that you implemented, write a paragraph or two about limitations / bottlenecks about the work. What could be improved? What seems to be some obvious issues with the existing works?

My model has a limitation on the spatial resolution in the encoder. Every feature in the encoder output is determined by a portion of the input (as shown in slide 18 of Topic 11), and thus, the feature does not take into account the entire image. An improvement of this would be to increase kernel sizes or use atrous convolutions (to avoid increasing run times) to increase the receptive field. 

Another bottleneck of the network is the slow training times. I used the Adam optimizer with a relatively small learning rate, which requires a large number of training epochs. This takes more time. Increasing the learning rate will increase the training times, but runs the risk of exploding gradients.

I also used resnet34, which is smaller than other resenet models (ex: resnet50). This may limit the number of useful higher dimensional features that the model could have learned.
